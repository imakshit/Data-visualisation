```python
# library & dataset
import seaborn as sns
df = sns.load_dataset('iris')
 
```

```python
# Make default density plot
sns.kdeplot(df['sepal_width'] , shade = True)
#sns.plt.show()
```

```python
# Make default density plot horizontal
sns.kdeplot(df['sepal_width'] , shade = True , vertical = True)
#sns.plt.show()
```

```python
# Large bandwidth
sns.kdeplot(df['sepal_width'], shade=True, bw=.5, color="blue")
#sns.plt.show()
 
# Narrower bandwidth
sns.kdeplot(df['sepal_width'], shade=True, bw=.05, color="olive")
#sns.plt.show()
```

```python
# plot of 2 variables
p1=sns.kdeplot(df['sepal_width'], shade=True, color="r")
p1=sns.kdeplot(df['sepal_length'], shade=True, color="b")
#sns.plt.show()
```
